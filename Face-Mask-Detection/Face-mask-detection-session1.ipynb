```python
import os
```

```python
from keras.preprocessing import image
```

```python
import cv2
```

```python
categories = ['with_mask','without_mask']
```

```python
data = []
for category in categories:
    path = os.path.join('train',category)

    label = categories.index(category)

    for file in os.listdir(path):

        img_path = os.path.join(path,file)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(224,224)) 

        data.append([img,label])

```

```python
len(data)
```

```python
import random
```

```python
random.shuffle(data)
```

```python
X = []
y = []

for features,label in data:
    X.append(features)
    y.append(label)

```

```python
len(X)
```

```python
len(y)
```

```python
import numpy as np
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X.shape
```

```python
y.shape
```

```python
X = X/255
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
vgg = VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model = Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```

```python
cap = cv2.VideoCapture(0)
```

```python
def detect_face_mask(img):
    y_pred_prob = model.predict(img.reshape(1, 224, 224, 3))
    threshold = 0.1  # Adjust threshold as needed
    y_pred_binary = (y_pred_prob > threshold).astype(int)
    return y_pred_binary[0][0]
```

```python
from sklearn.metrics import precision_score

# Example ground truth and predicted labels (replace with your actual data)
y_true = [0, 1, 2, 0, 1, 2]  # Ground truth labels (3-class problem)
y_pred = [0, 0, 1, 0, 0, 1]  # Predicted labels (we never predict '2')

# Calculate precision (macro average) with zero_division=1
precision_macro = precision_score(y_true, y_pred, average='macro', zero_division=1)

print("Precision (Macro):", precision_macro)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

cm = confusion_matrix(y_test, y_pred_classes)
print("Confusion Matrix:")
print(cm)

cr = classification_report(y_test, y_pred_classes)
print("Classification Report:")
print(cr)
```

```python
# Make predictions with threshold adjustment
threshold = 0.5
y_pred = (model.predict(X_test) > threshold).astype(int)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Compute classification report with zero division handling
cr = classification_report(y_test, y_pred, zero_division=1)
print("Classification Report:")
print(cr)
```

```python
def draw_label(img,text,pos,bg_color):
    text_size = cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    end_x = pos[0] + text_size[0][0] + 2
    end_y = pos[1] + text_size[0][1] - 2

    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,cv2.LINE_AA)
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()

    img = cv2.resize(frame,(224,224))

    y_pred = detect_face_mask(img)

    if y_pred == 0:
        draw_label(frame,"Mask",(30,30),(0,255,0))
    else:
        draw_label(frame,"No Mask",(30,30),(0,0,255))

    cv2.imshow("FaceMaskDetection", frame)

    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cap.release()
cv2.destroyAllWindows
```
